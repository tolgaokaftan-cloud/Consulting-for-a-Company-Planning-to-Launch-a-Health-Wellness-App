**2. Importing the Packages in our Pipeline**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
import matplotlib.pyplot as plt
```

We added all needed libraries and packages.

2.1 Data Collection

```python
df = pd.read_csv("/content/Smartphone_Usage_And_Addiction_Analysis_7500_Rows.csv")

print("Shape of my data:", df.shape)
df.head()
```

Now as we can see our data here. For the first steps, splitting test and train is crucial beforehand. **2.3 Data Preprocessing**

```python
df_train = df_train.dropna(subset=["addiction_level"])
df_test = df_test.dropna(subset=["addiction_level"])
```

In here we dropped the missing parts in our target column.

```python
drop_cols = ["transaction_id", "user_id", "addicted_label"]
df_train = df_train.drop(columns=drop_cols)
df_test = df_test.drop(columns=drop_cols)
```

This step is not cruical but we wanted to delete unnecessary part of our data.

```python
target = "addiction_level"

x_train = df_train.drop(target, axis=1)
y_train = df_train[target]

x_test = df_test.drop(target, axis=1)
y_test = df_test[target]
```

Features and target is splitted to be sure everything is going well as planned.

**2.4 Feature Engineering of Our Data**

```python
current_x_train_df = df_train.drop(target, axis=1)
current_x_test_df = df_test.drop(target, axis=1)

categorical_features = current_x_train_df.select_dtypes(include=["object"]).columns
numerical_features = current_x_train_df.select_dtypes(include=["int64", "float64"]).columns
encoder = OneHotEncoder(handle_unknown="ignore")
scaler = StandardScaler()

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features),
        ("num", StandardScaler(), numerical_features)
    ]
)

preprocessor.fit(current_x_train_df)

x_train = preprocessor.transform(current_x_train_df)
x_test = preprocessor.transform(current_x_test_df)

print("x_train size:", x_train.shape)
print("x_test size:", x_test.shape)
print("y_train size:", y_train.shape)
print("y_test size:", y_test.shape)
```

Encoding (numerical part of our features) and scaling (categorical part) to our data. Categorical data such as gender and behavioral attributes were converted into numerical data using one-hot encoding, which is a technique used for machine learning.
For the numerical data, standardization was used for scaling the data, ensuring all the data contributes equally to the model.

**2.5 Model Training**

For our problem, we want to try two different model training methods, also we want to test ensemble model. Then we want to see which one most suitable for our model.
Firstly, we started to imply SVM.

```python
svm_model = SVC()
svm_model.fit(x_train, y_train)

y_pred = svm_model.predict(x_test)

print(" Accuracy : ", accuracy_score(y_test, y_pred))

print("\n Classification Report :")
print(classification_report(y_test, y_pred))

print("\n Confusion Matrix :")
print(confusion_matrix(y_test, y_pred))
```

The baseline SVM model has an accuracy of about 57%. This shows that although there is some pattern recognition in the data that is being handled by the baseline SVM model, its score is hundred percent very low and needs to be improved.

The baseline SVM model is performing well in predicting the “Mild” class but is having difficulty in differentiating between “Moderate” and “Severe” classes. The confusion matrix is showing that there is a lot of confusion occurring in differentiating between “Moderate” and “Severe” classes. This is because there is not effective recognition of the difference between higher levels of addiction.

Now, we are gonna change our Model to see different outputs. Whether there are better results or not. We imply **Decision tree model** below.

```python
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(x_train, y_train)

y_pred = dt_model.predict(x_test)


print("Accuracy:", accuracy_score(y_test, y_pred))

print("\n  Report :")
print(classification_report(y_test, y_pred))

print("\n  Confusion Matrix :")
print(confusion_matrix(y_test, y_pred))
```

As we can see above, decision tree model is not accurate as SVM. Our accuracy was %57 percent and now it is %53. Model mixes up moderate and severe classes but it is working well for Mild. Even better then SVM.

Why our model is weak?
The Decision Tree model has performed relatively poorly due to the difficulty in identifying the complex and overlapping relationships between the “Moderate” and “Severe” classes. This is due to the highly flexible nature of the Decision Tree model, which makes it easily prone to overfitting the data and not performing well in generalizing the data, thus performing poorly in the separation of classes and the misclassification of the data between the classes. But we can use it while we are trying to ensemble the models.

Now, we are gonna try **random forest model** to move forward and then we will try to ensemble some models to see better results.

```python
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train, y_train)

y_pred = rf_model.predict(x_test)

print(" Accuracy :", accuracy_score(y_test, y_pred))

print("\n Report :")
print(classification_report(y_test, y_pred))

print("\n Confusion Matrix :")
print(confusion_matrix(y_test, y_pred))
```


Random Forest was used because of its ability to cope with complex nonlinear relationships and prevent overfitting through ensemble learning. The algorithm is best for behavioral data sets that have many features that interact with each other.

```python
param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5],
    "max_features": ["sqrt", "log2"]
}

grid = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid,
    cv=3,
    scoring="f1_macro",
    n_jobs=-1
)

grid.fit(x_train, y_train)

print("Best Params:", grid.best_params_)
```

Hyperparameter tuning was used for the Random Forest model to optimize its performance. Hyperparameters like the number of trees, depth of trees, and feature selection method were tuned to optimize bias-variance trade-off and improve the model's ability to generalize.

Why we choose n = 100, 200  depth = 10,20  min samples = 2,5 because: The values used for the hyperparameters were chosen to ensure that different levels of complexity are tested without making the process too slow. Since n_estimators  are commonly used values that yield stable results. The max_depth values were chosen to ensure that the complexity of the trees, both shallow and deeper, are used to prevent overfitting. The min_samples_split values are used to find a balance between a flexible model and a more general one. Lastly, the max_features options chosen, “sqrt” and “log2,” are commonly used to enhance the performance of the models by introducing randomness.

We could widen that but it would use too much cpu time.

```python
best_rf = grid.best_estimator_

y_pred = best_rf.predict(x_test)

print("Final Accuracy:", accuracy_score(y_test, y_pred))
print("\nFinal Report:")
print(classification_report(y_test, y_pred))
```

Because of the class imbalance in the dataset, accuracy alone is not sufficient to evaluate model performance. Therefore, the F1-scores are considered the primary evaluation metric, as it provides a balanced measure across all classes.

Finally we are gonna use different ensemble method; Which will be gradient boosting.

```python
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(x_train, y_train)

y_pred = gb_model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

In grradient boosting we saw the best results, if we test it more aggresively, I mean if we can train it in huge samples then we are gonna get better scores, cpu time will be more.

Among all the models, the best performance in the evaluation process has been obtained by the Gradient Boosting model. In comparison with other models, like SVM and Decision Tree, and the Random Forest model, the Gradient Boosting model has performed well in learning complex patterns in the data set. This is due to the reason that the boosting technique is used in the model, in which the focus is on learning from the previous errors.

The Gradient Boosting model has performed the best in comparison with other models due to the learning process of the model. In the Gradient Boosting model, the learning process is done sequentially. Unlike the Random Forest model, in the Gradient Boosting model, the trees are constructed one after the other, in which the focus is on learning the errors of the previous model. This helps the model in learning complex and overlapping classes, like the “Moderate” and “Severe” classes of addiction in the data set.

```python
models = ["SVM", "Decision Tree", "Random Forest", "Gradient Boosting"]
accuracy = [0.569, 0.53, 0.586, 0.60]
f1_scores = [0.60, 0.58, 0.62, 0.63]

x = range(len(models))

plt.figure(figsize=(12,5))

plt.bar(x, accuracy, width=0.4, label="Accuracy", align='center')
plt.bar([i + 0.4 for i in x], f1_scores, width=0.4, label="Macro F1")

plt.xticks([i + 0.2 for i in x], models)
plt.ylabel("Score")
plt.title("Model Performance Comparison")

plt.legend()
plt.show()
```
